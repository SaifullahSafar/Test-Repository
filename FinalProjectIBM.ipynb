```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
html_data = requests.get(url).text

soup = BeautifulSoup(html_data, "html.parser")

# پیدا کردن همه جدول‌ها
tables = soup.find_all("table")

# بررسی دقیق برای پیدا کردن جدول درست
for table in tables:
    if "Tesla Quarterly Revenue" in table.text:
        rows = table.find_all("tr")
        tesla_revenue = []

        for row in rows:
            cols = row.find_all("td")
            if len(cols) == 2:
                date = cols[0].text.strip()
                revenue = cols[1].text.strip()
                if revenue and revenue != "-":
                    tesla_revenue.append({"Date": date, "Revenue": revenue})
        break

# تبدیل به DataFrame
tesla_revenue_df = pd.DataFrame(tesla_revenue)
tesla_revenue_df.head()
```

```python
import dash
from dash import dcc, html
import plotly.graph_objs as go
import pandas as pd
import yfinance as yf
import requests
from bs4 import BeautifulSoup

# دریافت داده‌های سهام تسلا
tesla = yf.Ticker("TSLA")
tesla_history = tesla.history(period="max")
tesla_history.reset_index(inplace=True)

# دریافت درآمد تسلا
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

tables = soup.find_all("table")
tesla_revenue = []

for table in tables:
    if "Tesla Quarterly Revenue" in table.text:
        rows = table.find_all("tr")
        for row in rows:
            cols = row.find_all("td")
            if len(cols) == 2:
                date = cols[0].text.strip()
                revenue = cols[1].text.strip().replace("$", "").replace(",", "")
                if revenue and revenue != "-":
                    tesla_revenue.append({"Date": date, "Revenue": float(revenue)})
        break

tesla_revenue_df = pd.DataFrame(tesla_revenue)
tesla_revenue_df["Date"] = pd.to_datetime(tesla_revenue_df["Date"])

# ایجاد نمودار قیمت سهام
stock_fig = go.Figure([
    go.Scatter(x=tesla_history['Date'], y=tesla_history['Close'], mode='lines', name='Tesla Stock Price')
])
stock_fig.update_layout(title="Tesla Stock Price Over Time")

# ایجاد نمودار درآمد
revenue_fig = go.Figure([
    go.Bar(x=tesla_revenue_df['Date'], y=tesla_revenue_df['Revenue'], name='Tesla Revenue',
           marker=dict(color='rgb(0, 100, 200)'))
])
revenue_fig.update_layout(title="Tesla Revenue Over Time")

# ساخت داشبورد با Dash
app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("Tesla Stock vs Revenue Dashboard", style={'textAlign': 'center'}),
    dcc.Graph(figure=stock_fig, style={'height': '500px'}),
    dcc.Graph(figure=revenue_fig, style={'height': '500px'}),
])

if __name__ == "__main__":
    app.run(debug=True)
```
